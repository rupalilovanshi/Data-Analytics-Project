```python
data = pd.read_csv(r"C:\gilgigilichuu\owid-covid-data.csv")

print(data.head())
```

```python
data.columns
```

```python
continentwise=pd.pivot_table(data,values=["total_cases","new_cases","total_deaths","life_expectancy"],index="continent",aggfunc="max")
```

```python
continentwise["Recovery rate"]= continentwise["total_cases"]*100/continentwise["total_deaths"]
```

```python
continentwise["morality rate"]=continentwise["total_deaths"]*100/continentwise["life_expectancy"]
```

```python
continentwise=continentwise.sort_values(by="life_expectancy",ascending=False)
```

```python
continentwise.style.background_gradient(cmap="cubehelix")
```
